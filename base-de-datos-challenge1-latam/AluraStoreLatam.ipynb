# 3. Calificación promedio de la tienda

# 5. Envío promedio por tienda

```python
import pandas as pd
import matplotlib.pyplot as plt


def costo_envio():
    costox1 = len(tiendas['Costo de envío'])
    costox11 = tiendas['Costo de envío']

    totals = sum(costox11)
    total = totals/costox1

    print(f"El promedio de el costo de envio de la tienda 1: {total: .2f}")
    return total

def costo_envio2():
    costox2 = len(tiendas2['Costo de envío'])
    costox22 = tiendas2['Costo de envío']

    totals = sum(costox22)
    total = totals/costox2

    print(f"El promedio de el costo de envio de la tienda 2: {total: .2f}")
    return total

def costo_envio3():
    costox3 = len(tiendas3['Costo de envío'])
    costox33 = tiendas3['Costo de envío']

    totals = sum(costox33)
    total = totals/costox3

    print(f"El promedio de el costo de envio de la tienda 3: {total: .2f}")
    return total

def costo_envio4():
    costox4 = len(tiendas4['Costo de envío'])
    costox44 = tiendas4['Costo de envío']

    totals = sum(costox44)
    total = totals/costox4

    print(f"El promedio de el costo de envio de la tienda 4: {total: .2f}")
    return total



datos = [
    {"Tienda": "Tienda 1", "Costo Promedio": costo_envio()},
    {"Tienda": "Tienda 2", "Costo Promedio": costo_envio2()},
    {"Tienda": "Tienda 3", "Costo Promedio": costo_envio3()},
    {"Tienda": "Tienda 4", "Costo Promedio": costo_envio4()}
]
df = pd.DataFrame(datos)
print(df)

df.plot(kind='bar', x='Tienda', y='Costo Promedio', color='skyblue', figsize=(8, 5))


plt.title('Costo Promedio de Envío por Tienda')
plt.ylabel('Costo ($)')
plt.xticks(rotation=0) 
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()



```
